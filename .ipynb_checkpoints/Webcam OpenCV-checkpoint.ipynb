# 1. Import Dependencies

```python
import cv2
from matplotlib import pyplot as plt
```

# 2. Connecting to Your Webcam

```python
cap = cv2.VideoCapture(0)
```

```python
ret, frame = cap.read()
```

```python
print(ret)
```

```python
print(frame)
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def take_photo(): 
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('image-capture.jpg', frame)
    cap.release()
```

```python
take_photo()
```

```python
cap.release()
```

# 3. Rendering in Real Time

```python
cap = cv2.VideoCapture(0)
while cap.isOpened(): 
    ret, frame = cap.read()
    cv2.imshow('Webcam', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release()
cv2.destroyAllWindows()
```
