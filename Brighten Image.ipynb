```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('image-2.jpeg')
```

```python
alpha = 1.5
beta = 10
```

```python
adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
```

```python
plt.imshow(adjusted)
```

```python
cv2.imwrite('image-4.jpeg', adjusted)
```
